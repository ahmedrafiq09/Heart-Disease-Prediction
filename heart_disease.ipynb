```python
#Import All Libraries 
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("heart.csv")                               #Read CSV File 
df
```

```python
y = df.HeartDisease
y
```

```python
X = df.drop('HeartDisease', axis=1)                             #Drop the Labels of The Dataset
X
```

```python
#Use label Encoder to Conver The String Data To Int
le_Sex = LabelEncoder()
le_ChestPainType = LabelEncoder()
le_RestingECG = LabelEncoder()
le_ExerciseAngina = LabelEncoder()
le_ST_Slope = LabelEncoder()
```

```python
X['new_sex'] = le_Sex.fit_transform(X['Sex'])
X['new_ChestPainType'] = le_ChestPainType.fit_transform(X['ChestPainType'])
X['new_RestingECG'] = le_RestingECG.fit_transform(X['RestingECG'])
X['new_ExerciseAngina'] = le_ExerciseAngina.fit_transform(X['ExerciseAngina'])
X['new_ST_Slope'] = le_ST_Slope.fit_transform(X["ST_Slope"])
```

```python
X
```

```python
X = X.drop(['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'], axis=1)
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X.values, y, test_size=0.1)
```

```python
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

```python

model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[227, 230, 332, 0, 298, 0.0, 2219, 21, 2233, 0, 22]])
```
